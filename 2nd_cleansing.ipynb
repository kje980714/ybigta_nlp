```python
data_nn = data.dropna(axis=0).reset_index(drop=True)
data_nn.describe()
```

```python
len(data_nn)
```

```python
genres = {}
n = 0
for i in data_nn['genres']:
    if i == "[]":
        print(i)
    else:
        list = i.split("'")
        lst = []
        for item in list:
            if item.isalpha():
                if item not in genres:
                    genres[item] = 1
                else:
                    genres[item] += 1
                lst.append(item)
            data_nn[item] = 1
```

```python
for gen in genres.keys():
    data_nn[gen] = [0] * len(data_nn)
len(data_nn)
```

```python
n = 0
for gen in genres.keys():
    for i in data_nn['genres']:
        if gen in i:
            data_nn.loc[n,[gen]] = 1
            #data_nn[gen][n] = 1
        n += 1
    print(n == len(data_nn))
    n = 0
```
